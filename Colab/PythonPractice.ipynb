```python
# Pandas
# pip install pandas
# import pandas as pd 여기서 as는 불러온 라이브러리 별칭을 정할 때 사용

# Series : 1차원 형태의 데이터
# DataFrame : 2차원 형태의 데이터
# pd.Series() : 1차원 데이터 형태 생성
# pd.DataFrame() : 2차원 데이터프레임 생성

# 파일 불러오기
# pd.read_csv() : csv파일 불러오기
# pd.read_excel() : Excel파일 불러오기

import pandas as pd

data1 = pd.read_csv('/content/2014년 졸음운전 교통사고.csv', encoding='euc- kr')
data2 = pd.read_csv('/content/2015년 졸음운전 교통사고.csv', encoding='euc- kr')
data3 = pd.read_csv('/content/2016년 졸음운전 교통사고.csv', encoding='euc- kr')
```

```python
# 데이터 병합
# merge() : 같은 열을 기준으로 병합
# concat() : 단순 병합

data = pd.concat([data1,data2,data3])

# 인덱스 변경
# df.set_index(colums) : 특정 열을 인덱스로 변경(설정)
data.set_index('구분', inplace=True)      # inplace가 True일 경우 원본데이터 수정
data
```

```python
# data.index = pd.to_datetime(data.index, format="%Y년%m월")
# data.set_index([data.index.year, data.index.month], inplace=True)
data.index.names = ['년도', '월']
data
```

```python
# 년도별 평균 사망자
# df.groupby()
# 평균 : mean()
y = data.groupby('년도')['사망(명)'].mean()
y

# 월별 평균 사망자
x = data.groupby('월')['사망(명)'].mean()
x
```

```python
# loc : 열 이름으로 데이터 추출
# iloc : 열 위치로 데이터 추출

# 2016년 사고 대비 사망률
d_2016_sum = data.loc[2016]['사망(명)'].sum()
d_2016_ac = data.loc[2016]['사고(건)'].sum()
d_2016_rate = d_2016_sum / d_2016_ac
print(d_2016_rate)
```

```python
# 2014년도 사고 대비 부상자률
d_2014_sum = data.loc[2014]['부상(명)'].sum()
d_2014_ac = data.loc[2014]['사고(건)'].sum()
d_2014_rate = d_2014_sum / d_2014_ac
print(d_2014_rate)
```

```python
import seaborn as sns

# iris - 붓꽃
iris = sns.load_dataset('iris')
iris
```

```python
# sepal => 꽃받침. petal => 꽃잎, species => 꽃 종류
iris_data = iris.groupby('species').mean()
iris_data
```

```python
iris_data.plot(kind="bar", color=["skyblue", "black", "red", "yellow"])
plt.title("iris data")
plt.xlabel("Species")
plt.ylabel("Mean Value")
plt.yticks(rotation=0)
plt.xticks(rotation=0)
plt.show()
```

```python
data = sns.load_dataset('tips')
# total_bill => 계산 총액, tip => 팁 금액, sex => 손님 성별, smoker => 손님 흡연 여부
# day => 요일, time => 시간, size => 일행 수

# 1. 요일별 팁 평균
data_date_avertip = data.groupby('day')['tip'].mean()
print(data_date_avertip)

# 2. 성별에 따른 팁 평균 => 그래프
data_sex_avertip = data.groupby('sex')['tip'].mean()
data_sex_avertip.plot(kind="bar", color=["blue", "red"])
plt.title("Average Tip")
plt.xlabel("Sex")
plt.ylabel("Mean Value")
plt.yticks(rotation=0)
plt.xticks(rotation=0)
plt.show()

# 3. 흡연 여부에 따른 팁 비교
data_smoker_avertip = data.groupby('smoker')['tip'].mean()
data_smoker_avertip.plot(kind="bar", color=["gray", "black"])
plt.title("Average Tip")
plt.xlabel("Smoker")
plt.ylabel("Mean Value")
plt.yticks(rotation=0)
plt.xticks(rotation=0)
plt.show()
```
